```python
from dotenv import load_dotenv
load_dotenv()
import os
API_KEY_Weather = os.getenv("API_KEY_Weather")
```

Make a request for the current weather where you are born, or somewhere you've lived.

```python
import requests

url = f"https://api.tomorrow.io/v4/weather/realtime?location=48.137154,11.576124&apikey={API_KEY_Weather}"

headers = {
    "accept": "application/json"
}

response = requests.get(url, headers=headers)

data_muc = response.json()

print(data_muc)
```

Print out the country this location is in.

```python
country_lat = data_muc['location']['lat']
country_long = data_muc['location']['lon']

import geocoder
API_KEY_GoogleGeocode = os.getenv("API_KEY_GoogleGeocode")
g=geocoder.google([country_lat, country_long], method='reverse', key=API_KEY_GoogleGeocode)
g.country_long
```

What URL would I use to request a 3-day forecast at Heathrow?
f"https://api.tomorrow.io/v4/weather/forecast?location=51.470020,-0.454295&apikey={API_KEY_Weather}"
I can only have a 5-day forecast here.

```python
import requests

url = f"https://api.tomorrow.io/v4/weather/forecast?location=51.470020,-0.454295&apikey={API_KEY_Weather}"

headers = {
    "accept": "application/json"
}

response = requests.get(url, headers=headers)

data_london_forecast = response.json()

print(data_london_forecast)
```

```python
for days in data_london_forecast['timelines']['daily']:
    print(days['time'])
```

```python
for days in data_london_forecast['timelines']['daily']:
    print(f"On {days['time']}, the maximum temperature is {days['values']['temperatureMax']}°C.")
```

Print only the day with the highest maximum temperature.

```python
highest_day = ''
highest_temp = -999
for days in data_london_forecast['timelines']['daily']:
    if days['values']['temperatureMax'] > highest_temp:
        highest_temp = days['values']['temperatureMax']
        highest_day = days['time']
print(f"New highest temp day is {highest_day} with a temp of {highest_temp}.")
```
